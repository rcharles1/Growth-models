```python
import matplotlib.pyplot as plt
```

```python
# Define the exponential growth function
def minimum_wage(n):
    return 1.60 * (1.047)**n
```

```python
# Generate data for the x and y axes
years = range(1960, 1997)
wages = [minimum_wage(n-1960) for n in years]
```

```python
# Create the plot
plt.plot(years, wages)
plt.xlabel('Year')
plt.ylabel('Minimum Wage ($)')
plt.title('Predicted Growth of Minimum Wage Over Time')
```
